```python
import pandas as pd
from pathlib import Path

BASE_DIR = Path("../data")
bx_preprocessed = pd.read_csv(f"{BASE_DIR}/bookcrossing/Books Data with Category Language and Summary/Preprocessed_data.csv")
```

### select column

```python
bx_preprocessed = bx_preprocessed[["isbn", "book_title", "book_author", "year_of_publication", "publisher", "Language", "Category"]]
```

### clean string

```python
bx_preprocessed['Category'] = bx_preprocessed['Category'] \
    .str.strip("[]") \
    .str.replace("'", "", regex=False) \
    .str.replace('"', '', regex=False) \
    .str.strip()
```

### Keep only year

```python
bx_preprocessed['year_of_publication'] = pd.to_numeric(
    bx_preprocessed['year_of_publication'],
    errors='coerce'
)
```

### Remove duplicate books by ISBN

```python
before = len(bx_preprocessed)

bx_preprocessed = bx_preprocessed.drop_duplicates(subset=['isbn'])

after = len(bx_preprocessed)

print("Removed:", before - after)
```

### Convert "None" or empty values to real missing values

```python
bx_preprocessed = bx_preprocessed.replace(['None', 'nan', ''], None)
```

### Rename columns

```python
bx_preprocessed = bx_preprocessed.rename(columns={
    'book_title': 'title',
    'book_author': 'author',
    'publisher': 'publisher',
    'year_of_publication': 'published_year',
    'Category': 'genre',
    'Language': 'language'
})
```

### remove null data

```python
bx_preprocessed['genre'] = bx_preprocessed['genre'].replace('9', None)
bx_preprocessed['language'] = bx_preprocessed['language'].replace('9', None)
```
