## aggregation

```python
print(df_cross_ratings['ISBN'].duplicated().any())
print(df_amazon['Id'].duplicated().any())
print(df_7k['isbn10'].duplicated().any())
```

```python
# Aggregate each dataset to one row per book key before merging

def aggregate_for_merge(df, key_col, metrics, prefix):
    out = df.groupby(key_col, as_index=False).agg(metrics)

    # Flatten multi-level columns created by agg
    flat_cols = []
    for c in out.columns:
        if isinstance(c, tuple):
            base, stat = c
            if stat == '':
                flat_cols.append(base)
            else:
                flat_cols.append(f"{prefix}_{base}_{stat}")
        else:
            flat_cols.append(c)
    out.columns = flat_cols
    return out

amazon_agg = aggregate_for_merge(
    df_amazon,
    key_col='Id',
    metrics={'review/score': ['mean', 'count']},
    prefix='rating_amazon'
)

books7k_agg = aggregate_for_merge(
    df_7k,
    key_col='isbn10',
    metrics={'average_rating': ['mean']},
    prefix='rating_7k'
)

cross_agg = aggregate_for_merge(
    df_cross_ratings,
    key_col='ISBN',
    metrics={'Book-Rating': ['mean', 'count']},
    prefix='rating_book_crossing'
)

merge_2_agg = pd.merge(
    amazon_agg,
    books7k_agg,
    left_on='Id',
    right_on='isbn10',
    how='inner'
)
```
